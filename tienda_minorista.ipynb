```python
# quitando los 3 duplicados
df_sin_duplicados = df.drop_duplicates()
df = df_sin_duplicados
```

```python
#fecha de referencia
fecha_referencia = datetime(2024,8,1)
```

```python
rfm_df = df.groupby('Customer_ID').agg(
    Recencia = ('Date',lambda x: (fecha_referencia -x.max()).days),
    Frecuencia = ('Date','count'),
    Monto = ('Total_Amount','sum')
)
```

```python
rfm_df.shape
```

```python
rfm_df
```

```python
# uniendo el df anterior y el nuevo rfm
df2 = pd.merge(df,rfm_df,how='inner',on='Customer_ID')
df2.head(4)
```
